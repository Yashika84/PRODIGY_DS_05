# Traffic Analysis :

1) Most Common Cause of Accidents:
No Distancing was identified as the most frequent cause of road accidents in the dataset. This indicates that drivers are often not maintaining a safe distance from the vehicle in front of them, leading to frequent collisions. Such behavior is especially dangerous in high-speed or congested traffic conditions. The prominence of this cause underlines a critical issue in driver behavior and awareness. It reflects the need for stricter traffic regulations, better driver training, and awareness campaigns focused on maintaining proper distance while driving.
Closely following this are “Changing lanes to the right” and “Driving carelessly”, both pointing toward reckless or inattentive driving habits, which contribute heavily to road safety risks. These behaviors show a lack of discipline and road etiquette, especially in urban or high-density areas.

2) Accident Distribution by Day of the Week:
When analyzing accidents by day, Friday saw the highest number of reported accidents. This pattern may be associated with increased movement and traffic at the end of the workweek, fatigue, or a rush to wrap up work before the weekend.
Midweek days like Tuesday to Thursday also recorded consistently high accident numbers, likely due to regular office commuting and daytime congestion. Interestingly, Sundays and Mondays recorded fewer accidents, potentially due to lower traffic volumes or more cautious driving.

This distribution provides valuable input for traffic authorities to enforce safety measures more strictly on high-risk days.


3) Vehicle Type and Accident Involvement:
The most common vehicle types involved in accidents were Taxis, Automobiles, and Motorcycles. Their high accident rates can be attributed to :High frequency of use (public transport and two-wheelers),Lack of protective structures (especially for motorcycles)
On the other hand, heavy vehicles like trucks and buses were involved less frequently but could be linked to more severe or multi-vehicle collisions due to their size.

4) Severity Insights:
Slight Injuries made up the majority of recorded outcomes, suggesting that while accidents are frequent, most are not life-threatening.
Serious Injuries occurred in a smaller portion, often associated with higher-speed impacts or multiple vehicle involvement.
Fatal Injuries were relatively rare but often linked with causes like overspeeding, night-time poor lighting, or drunk driving.
This highlights that even if fatal accidents are fewer in number, they are highly preventable with focused enforcement and safety education.
